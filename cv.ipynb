```python
import cv2
from ultralytics import YOLO

# Load model
model = YOLO('E:/Data Mining/a6/best3.pt')

# Buka kamera lokal
cap = cv2.VideoCapture(0)  # Gunakan kamera lokal
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Prediksi dengan YOLOv8
    results = model(frame)
    annotated_frame = results[0].plot()  # Gambar bounding box pada frame

    # Tampilkan hasil
    cv2.imshow('Real-Time Detection', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
